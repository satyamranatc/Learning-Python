```python
from matplotlib import pyplot as plt
```

```python
plt.plot(["A","B","C","D"],[10,20,30,15], color = "r", marker = "+", linestyle = ":", linewidth = "1.5")
plt.plot(["A","B","C","D"],[15,10,25,15], color = "g", marker = "+", linestyle = ":", linewidth = "2.5")
plt.xlabel("Data")
plt.ylabel("Value")
plt.title("Plot")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 2)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 1)
plt.plot(x,y)

plt.show()
```

```python

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
plt.bar(["A","B","C","D"],[10,20,30,15], color = "r", width=.3)
plt.show()
```

```python
plt.barh(["A","B","C","D"],[10,20,30,15], color = "r", height=0.2)
plt.show()
```

```python
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "green", "b", "#4CAF50"]
plt.pie([10,20,30,15],startangle=90, labels=["A","B","C","D"], explode=myexplode, shadow=True, colors=mycolors)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels=mylabels)
plt.legend()
plt.show()
```
